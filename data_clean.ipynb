```python
# Подготовка датасета

import pandas as pd
df = pd.read_csv('Credit_OTP.csv', sep=';', encoding='cp1251', index_col='AGREEMENT_RK')
#df = pd.read_csv('Credit_OTP.csv', sep=';', encoding='cp1251')
```

```python
# Убираем предиктор с константой
df = df.drop(columns='DL_DOCUMENT_FL')
```

```python
# FAMILY_INCOME   as ordinal
#df['FAMILY_INCOME'] = df['FAMILY_INCOME'].astype('Categoric')
```

```python
# Убираем незначимые дроби и переводим в ИНТы

# PERSONAL_INCOME as INT           
df['PERSONAL_INCOME'] = df['PERSONAL_INCOME'].str.replace("[,.]\d{2}","",regex=True).astype('int32')

# CREDIT as int64
df['CREDIT'] = df['CREDIT'].str.replace("[,.]\d*","",regex=True).astype('int32')

# FST_PAYMENT as int64
df['FST_PAYMENT'] = df['FST_PAYMENT'].str.replace("[,.]\d*","",regex=True).astype('int32')

# LOAN_AVG_DLQ_AMT as int64
df['LOAN_AVG_DLQ_AMT'] = df['LOAN_AVG_DLQ_AMT'].str.replace("[,.]\d*","",regex=True).astype('int32')

# LOAN_MAX_DLQ_AMT as int64
df['LOAN_MAX_DLQ_AMT'] = df['LOAN_MAX_DLQ_AMT'].str.replace("[,.]\d*","",regex=True).astype('int32')
```

```python
#PREVIOUS_CARD_NUM_UTILIZED to INT: (NaN -> 0), 1, 2 
df['PREVIOUS_CARD_NUM_UTILIZED'] = df['PREVIOUS_CARD_NUM_UTILIZED'].fillna(0).astype('int32')
```
